$$A = \frac{1}{2}

x_3 & y_3 & 1 \\
\end{vmatrix}$$

```python
A = area([[0,0], [1,0], [0,1]])
```

ou como 

```python
v1 = (0,0);  v2 = (1,0);  v3 = (0,1)
vertices = [v1, v2, v3]
A = area(vertices)
```
